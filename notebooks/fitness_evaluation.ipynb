```python
with open('overlap_relative.pickle', 'rb') as char:
    overlap_relative = pickle.load(char)
```

```python
with open('candidate_eval.pickle', 'rb') as char:
    blob = pickle.load(char)
```

```python
candidate_set = blob[0]
hitlists = [blobi['qualified hits'] for blobi in candidate_set]
max_cover = candidate_set[0]['max coverage area']
```

```python
len(hitlists)
```

```python
face_areas = blob[1]['area']
```

```python
# greedy_solution = [1112, 661, 170, 1382, 490, 31, 627, 759, 1280, 1340, 175, 1094, 545, 377, 1444]
# greedy_solutions = [greedy_solution[0:i+1] for i in range(len(greedy_solution))]
greedy_solutions = [
    [1112, 661, 170, 1382, 490],
    [1112, 669, 1246, 300, 98]
]
# greedy_weighted = [300, 1094, 28, 1442, 524, 438, 1128, 771, 411, 1193, 490, 376, 1311, 1280, 606]
# greedy_weighteds = [greedy_weighted[0:i+1] for i in range(len(greedy_weighted))]
greedy_weighteds = [[300, 1094, 28, 1442, 524]]
GA_opti = [
    #[708, 299, 1386, 1066],
    [1384, 658, 360, 97, 1048]#,
    #[0, 524, 97, 1177, 842, 360],
    #[27, 1131, 225, 975, 712, 526, 70, 412, 1399, 835, 1381, 1078, 1279]
]
GA_opti_6 = [
    [343, 1095, 128, 1414, 627],
    [344, 626, 1064, 1361, 126],
    [0, 524, 97, 1177, 842, 360],
    [1045, 1272, 342, 126, 658]
]
GA_opti_6_2 = [[377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679], [377, 1180, 1080, 97, 679]]


# init random
random_size = 1000
population = []
for i in range(random_size):
    length = 5 #randint(1,15)
    indivi = [randint(0,len(candidate_set)-1) for j in range(length)]
    population.append(indivi)
```

```python
f = []
for k, one_indi in enumerate(population):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'random initial'
        }
    )

### append greedy tings
for k, one_indi in enumerate(greedy_solutions):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'greedy best first'
        }
    )
#########
#########
for k, one_indi in enumerate(greedy_weighteds):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'greedy weighted'
        }
    )
#####################################
#####################################
for k, one_indi in enumerate(GA_opti):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'GA 1'
        }
    )
for k, one_indi in enumerate(GA_opti_6):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'GA 6'
        }
    )
for k, one_indi in enumerate(GA_opti_6_2):
    _ = one_fit(
        individual=one_indi,
        overlap_rel=overlap_relative,
        qualified_hitlists_source=hitlists,
        arealist=face_areas,
        max_cov=max_cover)
    f.append(
        {
            'no': k,
            'fitness': _[0],
            'source': 'GA 6.2'
        }
    )

f = pd.DataFrame(f)
```
